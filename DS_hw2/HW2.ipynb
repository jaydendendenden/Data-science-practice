```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 為了處理方便，把 'train.csv' 和 'test.csv' 合併起來，'test.csv'的 CreditScore 欄位用 0 補起來。
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

from sklearn.preprocessing import StandardScaler
pd.set_option('display.float_format', lambda x: '%.2f' % x)
#norm_data = tempppp.copy()
scaler = StandardScaler()
scaler = scaler.fit(df.iloc[:,:21])
df.iloc[:,:21] = scaler.transform(df.iloc[:,:21])
#display(df)
scaler = StandardScaler()
scaler = scaler.fit(df_test.iloc[:,:21])
df_test.iloc[:,:21] = scaler.transform(df_test.iloc[:,:21])
display(df_test)
df_test['CreditScore'] = np.zeros((len(df_test),))

# 以 train_end_idx 作為 'train.csv' 和 'test.csv' 分界列，
train_end_idx = len(df)
print(train_end_idx)
df = pd.concat([df, df_test], sort=False)
#df = df.drop(columns=['col_2', 'col_13', 'col_17', 'col_20'])
df = df.drop(columns=['col_2', 'col_13', 'col_17','col_20'])

```

```python
from xgboost import XGBClassifier, cv, DMatrix
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=1000, stop=2000, num=10)]
max_depth = [int(x) for x in np.linspace(1, 5, num=5)]
max_depth.append(None)
learning_rate=[round(float(x),2) for x in np.linspace(start=0.01, stop=0.2, num=10)]
colsample_bytree =[round(float(x),2) for x in np.linspace(start=0.5, stop=1, num=5)]
gamma = [round(float(x),2) for x in np.linspace(start=0, stop=1, num=5)]

random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'learning_rate': learning_rate,
               'colsample_bytree': colsample_bytree,
               'gamma' : gamma}
X_train, y_train = df.drop(columns = ['CreditScore']).values[:train_end_idx, :], df['CreditScore'].values[:train_end_idx] 

from sklearn.utils.class_weight import compute_sample_weight
sample_weights = compute_sample_weight(
    class_weight='balanced',
    y=y_train #provide your own target name
)
xg4 = XGBClassifier(random_state=42,objective= 'multi:softmax',)

#Random search of parameters, using 3 fold cross validation, search across 100 different combinations, and use all available cores
xg_random = RandomizedSearchCV(estimator = xg4, param_distributions=random_grid,
                              n_iter=50, cv=4, verbose=10, random_state=42, n_jobs=4)
  
xg_random.fit(X_train,y_train)
xg_random.best_params_
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    df.drop(columns = ['CreditScore']).values[:train_end_idx, :],
    df['CreditScore'].values[:train_end_idx], test_size=0.2)

X_train, y_train = df.drop(columns = ['CreditScore']).values[:train_end_idx, :], df['CreditScore'].values[:train_end_idx] 
X_test = df.drop(columns = ['CreditScore']).values[train_end_idx:, :]

from sklearn.utils.class_weight import compute_sample_weight
sample_weights = compute_sample_weight(
    class_weight='balanced',
    y=y_train #provide your own target name
)

# create model instance
xgb = XGBClassifier(colsample_bytree=0.5, learning_rate=0.16, max_depth=5, n_estimators=2000,
                    # min_child_weight=1,
                    # subsample=0.8,
                    gamma=0,
                    early_stopping_rounds=10,
                    objective= 'multi:softmax',
                    seed=27)
# xgb = XGBClassifier(learning_rate =0.1,
#                     n_estimators=1000,
#                     max_depth=5,
#                     min_child_weight=1,
#                     gamma=0,
#                     subsample=0.8,
#                     colsample_bytree=0.8,
#                     objective= 'binary:logistic',
#                     nthread=4,
#                     scale_pos_weight=1,
#                     seed=27)
# fit model
xgb.fit(X_train, y_train, verbose=1,
        eval_set=[(X_train, y_train), (X_val, y_val)], sample_weight=sample_weights)
# xgb.fit(df.drop(columns = ['CreditScore']).values[:train_end_idx, :],
#     df['CreditScore'].values[:train_end_idx], verbose=1)

```

```python
from sklearn.metrics import accuracy_score, f1_score

ls = xgb.predict(X_val)
print('Accuracy: %f' % accuracy_score(y_val, ls))
print('f1-score: %f' % f1_score(y_val, ls, average='macro'))
ans_pred = xgb.predict(X_test)
df_sap = pd.DataFrame(ans_pred.astype(int), columns = ['CreditScore'])
df_sap.to_csv('ans1.csv',  index_label = 'Id')
```
